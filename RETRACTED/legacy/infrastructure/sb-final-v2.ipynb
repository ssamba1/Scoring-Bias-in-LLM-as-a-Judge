```python
PROBE_VARIANTS = {
    "rubric_order": ["control", "reversed", "random"],
    "score_id": ["numeric", "letter", "descriptive"],
    "reference_answer": ["none", "good", "poor"],
}

def make_prompt(item, probe_type, variant):
    i, r = item["instr"], item["resp"]
    if probe_type == "rubric_order":
        rubric = {"control":"Score from 1-5 (where 1 is worst, 5 is best)",
                  "reversed":"Score from 1-5 (where 1 is best, 5 is worst)",
                  "random":"Score from 1-5"}[variant]
        return f"Evaluate the following response.\n### Instruction: {i}\n### Response: {r}\n### {rubric}\n### Score:"
    elif probe_type == "score_id":
        rubric = {"numeric":"Score from 1-5",
                  "letter":"Score A-E (where A is best, E is worst)",
                  "descriptive":"Rate: Poor, Fair, Good, Very Good, Excellent"}[variant]
        return f"Evaluate the following response.\n### Instruction: {i}\n### Response: {r}\n### {rubric}\n### Score:"
    elif probe_type == "reference_answer":
        ref = "" if variant == "none" else "\n### Excellent response: This is a perfect answer." if variant == "good" else "\n### Poor response: This is wrong."
        return f"Evaluate the following response.\n### Instruction: {i}{ref}\n### Response: {r}\n### Score from 1-5 (where 1 is worst, 5 is best)\n### Score:"
print("Probes ready")

# DEBUG: Verify prompt format
test_prompt = make_prompt(ITEMS[0], 'rubric_order', 'control')
print(f"\nFirst 300 chars of prompt:\n{repr(test_prompt[:300])}")
```

```python
LETTER_MAP = {"A":5,"B":4,"C":3,"D":2,"E":1,"F":1}
DESCRIPTIVE_MAP = {"excellent":5,"very good":4,"good":3,"fair":2,"poor":1,"great":5,"average":3,"bad":1,"terrible":1,"perfect":5,
                   "outstanding":5,"exceptional":5,"satisfactory":3,"inadequate":1,"mediocre":2}

def parse_score(text, variant):
    """Parse score from text. Tries ALL methods as fallbacks."""
    if not text:
        return None
    t = text.strip().lower()
    numbers = re.findall(r"\b([1-5])\b", t)
    letter_match = re.search(r"\b([a-f])\b", t)
    desc_score = None
    for word, score in sorted(DESCRIPTIVE_MAP.items(), key=lambda x:-len(x[0])):
        if word in t:
            desc_score = float(score)
            break
    if variant == "numeric":
        if numbers: return float(numbers[0])
        if letter_match: return float(LETTER_MAP.get(letter_match.group(1).upper(), 0))
        return desc_score
    elif variant == "letter":
        if letter_match: return float(LETTER_MAP.get(letter_match.group(1).upper(), 0))
        if numbers: return float(numbers[0])
        return desc_score
    elif variant == "descriptive":
        if desc_score: return desc_score
        if numbers: return float(numbers[0])
        if letter_match: return float(LETTER_MAP.get(letter_match.group(1).upper(), 0))
        return None
    else:
        # rubric_order variants: try everything
        if numbers: return float(numbers[0])
        if letter_match: return float(LETTER_MAP.get(letter_match.group(1).upper(), 0))
        return desc_score

def score_model(model, tokenizer, prompt, probe_variant, device):
    """Score a single prompt, trying multiple parse strategies."""
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=1024).to(device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=20, temperature=0.0,
            do_sample=False, pad_token_id=tokenizer.eos_token_id, num_return_sequences=1)
    full = tokenizer.decode(outputs[0], skip_special_tokens=True)
    generated = full[len(prompt):].strip()
    
    # Strategy 1: extract after ### Score: marker
    if "### Score:" in generated:
        after_score = generated.split("### Score:")[-1].split("###")[0].strip()
        if after_score:
            s = parse_score(after_score, probe_variant)
            if s is not None and 1 <= s <= 5: return s
    
    # Strategy 2: search entire generated text
    s = parse_score(generated, probe_variant)
    if s is not None and 1 <= s <= 5: return s
    
    # Strategy 3: search FULL output for ### Score: followed by number
    full_numbers = re.findall(r"### Score:?\s*([1-5])", full)
    if full_numbers:
        return float(full_numbers[-1])
    
    return None

print("Parser ready - 3-strategy fallback")
```

```python
# === MODEL INVENTORY ===
# All IDs verified on HuggingFace
# Gemma2-9B requires license acceptance at: https://huggingface.co/google/gemma-2-9b

MODELS_7B = [
    ("mistralai/Mistral-7B-v0.3", "mistralai/Mistral-7B-Instruct-v0.3", "Mistral-7B-v0.3"),
    ("deepseek-ai/deepseek-llm-7b-base", "deepseek-ai/deepseek-llm-7b-chat", "DeepSeek-7B"),
    ("google/gemma-2-9b", "google/gemma-2-9b-it", "Gemma2-9B"),
    ("allenai/OLMo-7B-hf", "allenai/OLMo-7B-0724-Instruct-hf", "OLMo-7B"),
]
MODELS_SMALL = [
    ("Qwen/Qwen2.5-0.5B", "Qwen/Qwen2.5-0.5B-Instruct", "Qwen2.5-0.5B"),
    ("Qwen/Qwen2.5-3B", "Qwen/Qwen2.5-3B-Instruct", "Qwen2.5-3B"),
]

# Models that need chat template applied
DEEPSEEK_IDS = {"deepseek-ai/deepseek-llm-7b-chat"}

print(f"7B/9B: {len(MODELS_7B)} families (need 4-bit on T4 16GB)")
print(f"Small: {len(MODELS_SMALL)} families (fp16)")
print(f"Total models: {(len(MODELS_7B) + len(MODELS_SMALL)) * 2}")
print(f"\nNote: Gemma2-9B may fail if you haven't accepted its license on HF")
print(f"      Visit: https://huggingface.co/google/gemma-2-9b")
```

```python
all_results = {}
device = torch.device("cuda")

# Load checkpoint if exists (for resume after disconnect)
CHECKPOINT_FILE = "final_results.json"
if os.path.exists(CHECKPOINT_FILE):
    try:
        with open(CHECKPOINT_FILE) as f:
            all_results = json.load(f)
        print(f"Loaded checkpoint with {len(all_results)} models")
    except:
        print("Checkpoint load failed, starting fresh")
        all_results = {}

for families, is_7b in [(MODELS_7B, True), (MODELS_SMALL, False)]:
    for base_id, instruct_id, name in families:
        ids_to_run = [(base_id, name)]
        if instruct_id:
            ids_to_run.append((instruct_id, f"{name}-IT"))
        
        for model_id, model_name in ids_to_run:
            # Skip if already collected in checkpoint
            if model_name in all_results and all_results[model_name] is not None:
                if all(v is not None for v in all_results[model_name].values()):
                    print(f"\n⏩ Skipping {model_name} (already in checkpoint)")
                    continue
            
            print(f"\n{'='*60}\n{model_name}  ({model_id})\n{'='*60}")
            model = None
            try:
                if is_7b and USE_4BIT:
                    load_kwargs = {"quantization_config": BNB_CONFIG, "device_map": "auto"}
                    print("  Loading: 4-bit quantized")
                elif is_7b:
                    load_kwargs = {"torch_dtype": torch.float16, "device_map": "auto",
                                   "max_memory": {0: "14.5GiB", "cpu": "32GiB"}, "offload_folder": "offload"}
                    print("  Loading: fp16 + CPU offload")
                else:
                    load_kwargs = {"torch_dtype": torch.float16, "device_map": "auto"}
                    print("  Loading: fp16")

                tokenizer = AutoTokenizer.from_pretrained(model_id, token=os.environ["HF_TOKEN"])
                if tokenizer.pad_token is None:
                    tokenizer.pad_token = tokenizer.eos_token
                model = AutoModelForCausalLM.from_pretrained(model_id, **load_kwargs, token=os.environ["HF_TOKEN"])
                model.eval()
                pcount = sum(p.numel() for p in model.parameters())
                print(f"  Loaded: {pcount/1e6:.0f}M params | Device: {model.device}")
            except Exception as e:
                print(f"  FAILED to load: {e}")
                continue

            use_ct = model_id in DEEPSEEK_IDS
            scores = {pt: {} for pt in PROBE_VARIANTS}
            
            for pt, variants in PROBE_VARIANTS.items():
                for pv in variants:
                    item_scores = []
                    for idx, item in enumerate(ITEMS):
                        prompt = make_prompt(item, pt, pv)
                        if use_ct:
                            prompt = tokenizer.apply_chat_template(
                                [{"role": "user", "content": prompt}],
                                tokenize=False, add_generation_prompt=True
                            )
                        reps = []
                        for _ in range(3):
                            try:
                                s = score_model(model, tokenizer, prompt, pv, device)
                                if s is not None and 1 <= s <= 5:
                                    reps.append(s)
                            except:
                                pass
                        if reps:
                            item_scores.append(np.mean(reps))
                        if (idx + 1) % 25 == 0:
                            print(f"  [{pt}/{pv}] {idx+1}/{len(ITEMS)}  ({len(item_scores)} valid scores)", flush=True)
                    
                    scores[pt][pv] = {"mean": float(np.mean(item_scores)) if item_scores else None}
                    
                    # Debug: show raw output if 0 scored
                    if len(item_scores) == 0:
                        prompt2 = make_prompt(ITEMS[0], pt, pv)
                        if use_ct:
                            prompt2 = tokenizer.apply_chat_template(
                                [{"role": "user", "content": prompt2}],
                                tokenize=False, add_generation_prompt=True
                            )
                        inputs = tokenizer(prompt2, return_tensors="pt", truncation=True, max_length=1024).to(device)
                        with torch.no_grad():
                            outputs = model.generate(**inputs, max_new_tokens=30, temperature=0.0,
                                do_sample=False, pad_token_id=tokenizer.eos_token_id)
                        debug_out = tokenizer.decode(outputs[0], skip_special_tokens=True)
                        gen_part = debug_out[len(prompt2):].strip()[:200]
                        print(f"  ⚠️  Zero scored — model output: \"{gen_part}\"")

            # Compute deltas
            result = {}
            for pt, pv in PROBE_VARIANTS.items():
                means = [scores[pt][v]["mean"] for v in pv if scores[pt][v]["mean"] is not None]
                result[pt] = round(max(means) - min(means), 2) if len(means) >= 2 else None
            all_results[model_name] = result
            print(f"  ✅ DONE: {result}")
            
            # Checkpoint after each model
            with open(CHECKPOINT_FILE, "w") as f:
                json.dump(all_results, f, indent=2)
            print(f"  💾 Checkpoint saved ({len(all_results)} models)")
            
            # Clean up
            del model
            gc.collect()
            torch.cuda.empty_cache()
            time.sleep(3)

with open(CHECKPOINT_FILE, "w") as f:
    json.dump(all_results, f, indent=2)
print(f"\n{'='*60}\n🎉 FINAL: {len(all_results)} models collected\n{'='*60}")
```
